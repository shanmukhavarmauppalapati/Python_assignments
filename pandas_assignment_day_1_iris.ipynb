##### 5. Write a Pandas program to create a subtotal of "PetalLengthCm" against Species from the dataframe.

```python
df['Species'].unique()
```

```python
### code here
v = df.groupby("Species")
```

```python
v.groups # in speices we have different data
```

```python
A=v.get_group("Iris-setosa")
B=v.get_group("Iris-versicolor")
c=v.get_group("Iris-virginica")
print("Iris-virginica:",c["PetalLengthCm"].sum())
print("Iris-versicolor:",B["PetalLengthCm"].sum())
print("Iris-setosa:",A["PetalLengthCm"].sum())
```
