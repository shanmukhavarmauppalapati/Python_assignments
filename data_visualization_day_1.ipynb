###### 1. load the iris.csv dataset.

```python
#We import the libraries what are used for this dataset.
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
plt.style.use('seaborn-bright')
```

```python
sns.get_dataset_names()
```

```python
#Here we load the dataset from seaborn 
iris = sns.load_dataset('iris')
```

```python
iris.head() #we collect the data
```

```python
iris.info() #here we check the data where it is clean dataset or not 
```

```python
iris.isnull().sum()
```

#### 1. create a bar plot for count the frequency of the three species.

```python
sns.countplot(x='species',data=iris)
plt.title("Frequency of three Species Count")
plt.show()
```

##### We create a count plot to check the frquency of the species and we see that all the three species have the same frequency.

##### 2. Create a pie chart to show the percentage of species.

```python
ax=plt.subplots(1,1,figsize=(10,8))
iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Percentage of Species")
plt.show()
```

###### Here we can see that all the species having the same percentage of values.

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='reg')
plt.title("Relation btw two variables")
plt.show()
```

##### Here we see the relation btw two sepal length and sepal width by using the Jointplot and their distributions on the same plot.

```python
sns.jointplot(x='petal_length',y='petal_width',data=iris,kind='reg')
plt.title("Relation btw two variables")
plt.show()
```

##### Here we see the relation btw two petal length and petal width by using the Jointplot and their distributions on the same plot.

##### 4. Create a scatter plot and take hue = "Species"

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
plt.title("Scatter plot for Iris")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.title("Scatter plot for Iris")
plt.show()
```

##### Here we see the relation btw two sepal length ,petal length,petal width and sepal width by using the scatterplot and we use the 'hue' to specifically show that the relation.

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
plt.figure(figsize=(15,10))
box_data = iris #variable representing the data array
box_target = iris.species #variable representing the labels array
sns.boxplot(data = box_data,width=0.5,fliersize=5)
```

##### Here we use the Box plot to individual show that the variance of the columns by comparing species to all the columns of the iris.

###### 6. Create a Strip plot 

```python
plt.figure(figsize=(10,8))
sns.stripplot(x='sepal_width',y='sepal_length',hue="species",data=iris,jitter=0.1)
plt.xticks(rotation=90)
plt.show()
```

###### 7. Create a Violin Plot.

```python
plt.figure(figsize=(10,10))
sns.catplot(x='species',y='sepal_length',data=iris,kind='violin')
plt.show()
```

>- In the violin we can observe the combination of species and their sepallength.
>- The white dot tells about the median of the species.

##### 8. Create a pair chart.

```python
sns.pairplot(iris,hue='species') #PAIR PLOT IS USED ONLY TO THE FEW NUMERICALS VALUES ARE TO BE ANALYSED
plt.show()
```

##### 9. Create a Heat Map

```python
plt.figure(figsize=(10,5))
sns.heatmap(iris.corr(),annot=True)
plt.show()
```

>- Heatmap is not constrained by no.of numerical variables like the pairplot.
>- It is faster than the pairplot.
>- very useful plot to check correlation btw any two numerical variables.
