##### 1. Write a Python program to display a bar chart of the total bill of Gender and write the intuition about the plot.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.show()
```

#### In this plot we observe that the Gender of the male are high when compared to female.

###### 2. Write a Python program to display a box plot of the total bill of day and hue = "smoker" and write the intuition about the plot.

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
plt.show()
```

>- We observe that the saturaday have the high outliers when compared by others.
>- And the sunday most of the people smoke who are come to the restaurants.

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
plt.figure(figsize=(10,10))
sns.catplot(x='day',y='total_bill',data=tips,kind='violin',hue='smoker')
plt.show()
```

>- violin plot is nothing but boxplot.
>- in this we can see the data distribution and combination plot.
>- the white dot tell about the median of the value.
>- density and frequency are also seen outside of the median.
>- hue is used to separate the data further using the day category
>- setting split=True will draw half of a violin for each level. This can make it easier to directly compare the distributions.

##### 4. Write a Python program to display a strip plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
plt.figure(figsize=(10,8))
sns.stripplot(x='day', y='total_bill', data=tips,
              jitter=True, hue='sex', dodge=True)
plt.show()
```

>- One problem with strip plot is that you cant really tell which points are stacked on top of each other and hence we use the jitter parameter to add some random days.
>- jitter parameter is used to add an amount of jitter (only along the categorical axis) which can be useful when you have many points and they overlap so that it is easier to see the distribution.
>- hue is used to provide an additional categorical separation
>- setting split=True is used to draw separate strip plots based on the category specified by the hue parameter

##### 5. Write a Python program to display heat map of dataframe and write the intuition about the plot.

```python
plt.figure(figsize=(10,5))
sns.heatmap(tips.corr(),annot=True)
plt.show()
```

>- Heatmap is not constrained by no.of numerical variables like the pairplot.
>- It is faster than the pairplot.
>- very useful plot to check correlation btw any two numerical variables.

###### 1. What is Histogram chart ?

A histogram is a display of statistical information that uses rectangles to show the frequency of data items in successive numerical intervals of equal size. In the most common form of histogram, the independent variable is plotted along the horizontal axis and the dependent variable is plotted along the vertical axis.

##### 2. What is box plot ?

A box plot is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum. The term "box plot" comes from the fact that the graph looks like a rectangle with lines extending from the top and bottom.

##### 3. What are the plots we can use for univeriate analysis. 


A univariate plot shows the data and summarizes its distribution. A dot plot, also known as a strip plot, shows the individual observations. A box plot shows the five-number summary of the data  the minimum, first quartile, median, third quartile, and maximum

##### 4. What are the plots we can use for bivariate analysis.

The scatter diagram or scatter plot is the workhorse bivariate plot, and is probably the plot type that is most frequently generated in practice (which is why it is the default plot method in R).

###### 5. What are plot we can use for multivariate analysis.

When you have a bivariate data, you can easily visualize the relationship between the two variables by plotting a simple scatter plot. For a data set containing three continuous variables, you can create a 3d scatter plot
